# Importando as Bibliotecas


```python
import numpy as pd
import pandas as pd

import matplotlib.pyplot as plt
%matplotlib inline 

import seaborn as sns
sns.set_style('darkgrid')
```

```python
voos = sns.load_dataset('flights')
voos.head()
```

# Gráfico de Linha 


```python
voos_maio = voos[voos['month'] == 'May']
sns.lineplot(data=voos_maio, x='year', y='passengers')

plt.title('Total de Passageiros na Viagem ocorridas no mês de Maio ao longo dos anos')
plt.ylabel('Passageiros')
plt.xlabel('Anos')
```

# Gráfico com todos os meses

```python
sns.lineplot(data=voos, x='year', y='passengers', hue='month')

plt.title('Total de passageiros divididos pelo mês da viagem')
plt.ylabel('Passageiros')
plt.xlabel('Ano')
```

# Heatmap - Mapa de calor

```python
#voos = voos.pivot('month','year','passengers')

sns.heatmap(voos, cmap='YlGnBu')

plt.title('Mapa de calor com o número de passageiros e seus respectivos ano e mês de viagem')
plt.ylabel('')
plt.xlabel('')
```
